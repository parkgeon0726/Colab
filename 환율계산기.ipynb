5. API vs 웹 크롤링 비교

데이터 제공 방식 - API:서비스에서 공식 제공 , 웹 크롤링:웹페이지에서 직접 추출

데이터 형식 - API:JSON, XML 등 구조화된 형식 , 웹 크롤링:HTML에서 파싱 필요

안정성 - API:매우 안정적 (공식 지원) , 웹 크롤링:웹사이트 변경 시 오류 발생 가능

법적 문제 - API:안전함 (공식 허가) , 웹 크롤링:저작권 문제 발생 가능

사용 제한 - API:호출 횟수 제한 있음 , 웹 크롤링:제한 없지만 서버 부하 고려

비용 - API:무료~유료 (사용량에 따라) , 웹 크롤링:대부분 무료


6. API 사용이 좋은 경우:공식 API가 제공될 때 , 실시간 데이터가 필요할 때 , 구조화된 데이터가 필요할 때 , 장기적으로 사용할 서비스일 때

7. 웹 크롤링을 고려해야 하는 경우: API가 제공되지 않을 때 , 매우 특수한 데이터가 필요할 때 , 일회성 데이터 수집일 때 , 비용을 절약해야 할 때

ExchangeRate-API 사용 (무료, API 키 불필요)

실시간 환율 정보 자동 업데이트

사용자 친화적인 메뉴 인터페이스

소수점 5자리까지 정확한 계산 지원

```python
import requests
import json

class CurrencyCalculator:
    def __init__(self):
        self.exchange_rates = {}
        self.base_url = "https://api.exchangerate-api.com/v4/latest/USD"

    def get_exchange_rates(self):
        """
        실시간 환율 정보를 가져오는 함수
        USD를 기준으로 다른 통화의 환율을 가져옴
        """
        try:
            response = requests.get(self.base_url)
            if response.status_code == 200:
                data = response.json()
                self.exchange_rates = data['rates']
                print("✅ 환율 정보를 성공적으로 업데이트했습니다!")
                print(f"📅 업데이트 시간: {data.get('date', 'N/A')}")
                return True
            else:
                print("❌ 환율 정보를 가져올 수 없습니다.")
                return False
        except Exception as e:
            print(f"❌ 오류 발생: {e}")
            return False

    def usd_to_krw(self, usd_amount):
        """
        달러를 원화로 변환
        """
        if 'KRW' in self.exchange_rates:
            krw_amount = usd_amount * self.exchange_rates['KRW']
            return krw_amount
        else:
            print("❌ 원화 환율 정보를 찾을 수 없습니다.")
            return None

    def jpy_to_krw(self, jpy_amount):
        """
        엔화를 원화로 변환
        JPY는 100엔 단위로 계산됨을 고려
        """
        if 'KRW' in self.exchange_rates and 'JPY' in self.exchange_rates:
            # JPY -> USD -> KRW 순서로 변환
            usd_amount = jpy_amount / self.exchange_rates['JPY']
            krw_amount = usd_amount * self.exchange_rates['KRW']
            return krw_amount
        else:
            print("❌ 엔화 또는 원화 환율 정보를 찾을 수 없습니다.")
            return None

    def display_current_rates(self):
        """
        현재 환율 정보 표시
        """
        if self.exchange_rates:
            usd_to_krw = self.exchange_rates.get('KRW', 0)
            jpy_to_usd = self.exchange_rates.get('JPY', 0)

            print("\n" + "="*50)
            print("💱 현재 환율 정보")
            print("="*50)
            print(f"🇺🇸 1 USD = {usd_to_krw:.2f} KRW")
            if jpy_to_usd > 0:
                jpy_to_krw_rate = usd_to_krw / jpy_to_usd
                print(f"🇯🇵 1 JPY = {jpy_to_krw_rate:.4f} KRW")
                print(f"🇯🇵 100 JPY = {jpy_to_krw_rate * 100:.2f} KRW")
            print("="*50)

    def run_calculator(self):
        """
        메인 계산기 실행 함수
        """
        print("💰 환율 계산기에 오신 것을 환영합니다!")
        print("🔄 환율 정보를 불러오는 중...")

        if not self.get_exchange_rates():
            print("환율 정보를 가져올 수 없어 프로그램을 종료합니다.")
            return

        self.display_current_rates()

        while True:
            print("\n" + "="*50)
            print("📋 메뉴를 선택하세요:")
            print("1. 달러(USD) → 원화(KRW) 변환")
            print("2. 엔화(JPY) → 원화(KRW) 변환")
            print("3. 현재 환율 정보 보기")
            print("4. 환율 정보 업데이트")
            print("5. 종료")
            print("="*50)

            try:
                choice = input("선택 (1-5): ").strip()

                if choice == '1':
                    self.handle_usd_conversion()
                elif choice == '2':
                    self.handle_jpy_conversion()
                elif choice == '3':
                    self.display_current_rates()
                elif choice == '4':
                    print("🔄 환율 정보를 업데이트하는 중...")
                    self.get_exchange_rates()
                elif choice == '5':
                    print("👋 환율 계산기를 종료합니다. 감사합니다!")
                    break
                else:
                    print("❌ 잘못된 선택입니다. 1-5 중에서 선택해주세요.")

            except KeyboardInterrupt:
                print("\n👋 프로그램을 종료합니다.")
                break
            except Exception as e:
                print(f"❌ 오류가 발생했습니다: {e}")

    def handle_usd_conversion(self):
        """
        달러 변환 처리
        """
        try:
            print("\n💵 달러(USD) → 원화(KRW) 변환")
            usd_input = input("변환할 달러 금액을 입력하세요 (소수점 5자리까지 지원): ")
            usd_amount = float(usd_input)

            if usd_amount < 0:
                print("❌ 음수는 입력할 수 없습니다.")
                return

            krw_result = self.usd_to_krw(usd_amount)
            if krw_result is not None:
                print(f"\n✅ 변환 결과:")
                print(f"💵 {usd_amount:,.5f} USD = 💰 {krw_result:,.2f} KRW")

        except ValueError:
            print("❌ 올바른 숫자를 입력해주세요.")
        except Exception as e:
            print(f"❌ 변환 중 오류가 발생했습니다: {e}")

    def handle_jpy_conversion(self):
        """
        엔화 변환 처리
        """
        try:
            print("\n💴 엔화(JPY) → 원화(KRW) 변환")
            jpy_input = input("변환할 엔화 금액을 입력하세요 (소수점 5자리까지 지원): ")
            jpy_amount = float(jpy_input)

            if jpy_amount < 0:
                print("❌ 음수는 입력할 수 없습니다.")
                return

            krw_result = self.jpy_to_krw(jpy_amount)
            if krw_result is not None:
                print(f"\n✅ 변환 결과:")
                print(f"💴 {jpy_amount:,.5f} JPY = 💰 {krw_result:,.2f} KRW")

        except ValueError:
            print("❌ 올바른 숫자를 입력해주세요.")
        except Exception as e:
            print(f"❌ 변환 중 오류가 발생했습니다: {e}")

# 간단 사용 예제
def simple_converter():
    """
    간단한 원라이너 변환 함수들
    """
    print("\n" + "="*50)
    print("🚀 간단 변환 함수 예제")
    print("="*50)

    def quick_usd_to_krw(usd_amount):
        """USD를 KRW로 빠르게 변환"""
        try:
            response = requests.get("https://api.exchangerate-api.com/v4/latest/USD")
            if response.status_code == 200:
                rate = response.json()['rates']['KRW']
                return usd_amount * rate
            return None
        except:
            return None

    def quick_jpy_to_krw(jpy_amount):
        """JPY를 KRW로 빠르게 변환"""
        try:
            response = requests.get("https://api.exchangerate-api.com/v4/latest/USD")
            if response.status_code == 200:
                rates = response.json()['rates']
                usd_amount = jpy_amount / rates['JPY']
                return usd_amount * rates['KRW']
            return None
        except:
            return None

    # 예제 실행
    print("예제 1: 100 USD를 KRW로 변환")
    result = quick_usd_to_krw(100)
    if result:
        print(f"100 USD = {result:,.2f} KRW")

    print("\n예제 2: 10000 JPY를 KRW로 변환")
    result = quick_jpy_to_krw(10000)
    if result:
        print(f"10000 JPY = {result:,.2f} KRW")

# 메인 실행부
if __name__ == "__main__":
    calculator = CurrencyCalculator()
    calculator.run_calculator()

    # 간단 변환 함수 예제도 보여주기
    # simple_converter()
```

ExchangeRate-API 사용으로 안정적인 데이터 제공

최소한의 코드로 핵심 기능 구현

실시간 환율 정보 새로고침 기능

API 호출 오류 처리 포함

```python
# ExchangeRate-API 사용 버전 (무료, API 키 불필요)
import requests

def get_exchange_rates():
    """
    ExchangeRate-API에서 실시간 환율 정보 가져오기
    """
    try:
        url = "https://api.exchangerate-api.com/v4/latest/USD"
        response = requests.get(url)

        if response.status_code == 200:
            return response.json()['rates']
        return None
    except Exception as e:
        print(f"API 호출 오류: {e}")
        return None

def currency_converter():
    """
    API 기반 환율 계산기
    """
    print("🌍 실시간 환율 계산기 (API 기반)")
    print("="*45)

    # 환율 정보 가져오기
    rates = get_exchange_rates()
    if not rates:
        print("환율 정보를 가져올 수 없습니다.")
        return

    usd_krw = rates['KRW']
    jpy_usd = rates['JPY']
    jpy_krw = usd_krw / jpy_usd  # 1 JPY = ? KRW

    print(f"\n💱 현재 환율:")
    print(f"🇺🇸 1 USD = {usd_krw:,.2f} KRW")
    print(f"🇯🇵 1 JPY = {jpy_krw:.4f} KRW")
    print(f"🇯🇵 100 JPY = {jpy_krw*100:.2f} KRW")

    while True:
        print("\n" + "="*45)
        print("1️⃣ 달러 → 원화")
        print("2️⃣ 엔화 → 원화")
        print("3️⃣ 환율 정보 새로고침")
        print("4️⃣ 종료")

        choice = input("\n메뉴 선택: ")

        if choice == '1':
            try:
                amount = float(input("달러 금액 입력 (소수점 5자리까지): "))
                result = amount * usd_krw
                print(f"\n💵 ${amount:,.5f} = 💰 {result:,.2f}원")
            except ValueError:
                print("❌ 유효한 숫자를 입력하세요.")

        elif choice == '2':
            try:
                amount = float(input("엔화 금액 입력 (소수점 5자리까지): "))
                result = amount * jpy_krw
                print(f"\n💴 ¥{amount:,.5f} = 💰 {result:,.2f}원")
            except ValueError:
                print("❌ 유효한 숫자를 입력하세요.")

        elif choice == '3':
            print("🔄 환율 정보 업데이트 중...")
            rates = get_exchange_rates()
            if rates:
                usd_krw = rates['KRW']
                jpy_usd = rates['JPY']
                jpy_krw = usd_krw / jpy_usd
                print("✅ 환율 정보가 업데이트되었습니다!")
                print(f"🇺🇸 1 USD = {usd_krw:,.2f} KRW")
                print(f"🇯🇵 1 JPY = {jpy_krw:.4f} KRW")
            else:
                print("❌ 환율 정보 업데이트 실패")

        elif choice == '4':
            print("👋 계산기를 종료합니다.")
            break
        else:
            print("❌ 1-4 중에서 선택하세요.")

if __name__ == "__main__":
    currency_converter()
```

파싱(parsing)은 “데이터나 문장(텍스트)을 분석해서 의미 있는 단위로 분해하고 구조화하는 작업"을 말합

프로그래밍에서는 주로 웹 페이지, 파일, API 응답 등에서 받은 원시(raw) 데이터를 프로그래머가 다루기 편한 형태(객체, 리스트, 딕셔너리 등)로 변환할 때 사용함

왜 파싱이 필요함?
1. 데이터 추출
웹 페이지의 HTML 코드에서 뉴스 제목, 이미지 URL, 표 안의 값 등을 뽑아 내려면 HTML 태그와 그 구조를 이해해야 함
2. 구조화된 정보 활용
JSON이나 XML 형식으로 전달된 API 응답을 파싱해 객체로 변환하면 손쉽게 각 필드에 접근할 수 있음
3. 자동화 및 분석
대량의 문서나 로그 파일을 파싱해 필요한 정보만 추출하면 분석·처리 작업을 자동화할 수 있음

언제 파싱을 사용함?
1. 웹 크롤러 개발 시: 여러 웹페이지를 방문해 필요한 정보만 모으고자 할 때
2. API 응답 처리 시: JSON, XML 등으로 받은 응답을 연산하거나 저장할 때
3. 로그 파일 분석 시: 로그 텍스트 안에서 특정 패턴(날짜, 에러 메시지 등)만 추출할 때
4. 데이터 전처리 단계: 머신러닝이나 데이터 분석을 위해 텍스트 데이터를 정리할 때
