```python
import requests
import json

class CurrencyCalculator:
    def __init__(self):
        self.exchange_rates = {}
        self.base_url = "https://api.exchangerate-api.com/v4/latest/USD"

    def get_exchange_rates(self):
        """
        ì‹¤ì‹œê°„ í™˜ìœ¨ ì •ë³´ë¥¼ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜
        USDë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‹¤ë¥¸ í†µí™”ì˜ í™˜ìœ¨ì„ ê°€ì ¸ì˜´
        """
        try:
            response = requests.get(self.base_url)
            if response.status_code == 200:
                data = response.json()
                self.exchange_rates = data['rates']
                print("âœ… í™˜ìœ¨ ì •ë³´ë¥¼ ì„±ê³µì ìœ¼ë¡œ ì—…ë°ì´íŠ¸í–ˆìŠµë‹ˆë‹¤!")
                print(f"ğŸ“… ì—…ë°ì´íŠ¸ ì‹œê°„: {data.get('date', 'N/A')}")
                return True
            else:
                print("âŒ í™˜ìœ¨ ì •ë³´ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
                return False
        except Exception as e:
            print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
            return False

    def usd_to_krw(self, usd_amount):
        """
        ë‹¬ëŸ¬ë¥¼ ì›í™”ë¡œ ë³€í™˜
        """
        if 'KRW' in self.exchange_rates:
            krw_amount = usd_amount * self.exchange_rates['KRW']
            return krw_amount
        else:
            print("âŒ ì›í™” í™˜ìœ¨ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return None

    def jpy_to_krw(self, jpy_amount):
        """
        ì—”í™”ë¥¼ ì›í™”ë¡œ ë³€í™˜
        JPYëŠ” 100ì—” ë‹¨ìœ„ë¡œ ê³„ì‚°ë¨ì„ ê³ ë ¤
        """
        if 'KRW' in self.exchange_rates and 'JPY' in self.exchange_rates:
            # JPY -> USD -> KRW ìˆœì„œë¡œ ë³€í™˜
            usd_amount = jpy_amount / self.exchange_rates['JPY']
            krw_amount = usd_amount * self.exchange_rates['KRW']
            return krw_amount
        else:
            print("âŒ ì—”í™” ë˜ëŠ” ì›í™” í™˜ìœ¨ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
            return None

    def display_current_rates(self):
        """
        í˜„ì¬ í™˜ìœ¨ ì •ë³´ í‘œì‹œ
        """
        if self.exchange_rates:
            usd_to_krw = self.exchange_rates.get('KRW', 0)
            jpy_to_usd = self.exchange_rates.get('JPY', 0)

            print("\n" + "="*50)
            print("ğŸ’± í˜„ì¬ í™˜ìœ¨ ì •ë³´")
            print("="*50)
            print(f"ğŸ‡ºğŸ‡¸ 1 USD = {usd_to_krw:.2f} KRW")
            if jpy_to_usd > 0:
                jpy_to_krw_rate = usd_to_krw / jpy_to_usd
                print(f"ğŸ‡¯ğŸ‡µ 1 JPY = {jpy_to_krw_rate:.4f} KRW")
                print(f"ğŸ‡¯ğŸ‡µ 100 JPY = {jpy_to_krw_rate * 100:.2f} KRW")
            print("="*50)

    def run_calculator(self):
        """
        ë©”ì¸ ê³„ì‚°ê¸° ì‹¤í–‰ í•¨ìˆ˜
        """
        print("ğŸ’° í™˜ìœ¨ ê³„ì‚°ê¸°ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!")
        print("ğŸ”„ í™˜ìœ¨ ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¤ëŠ” ì¤‘...")

        if not self.get_exchange_rates():
            print("í™˜ìœ¨ ì •ë³´ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ì—†ì–´ í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            return

        self.display_current_rates()

        while True:
            print("\n" + "="*50)
            print("ğŸ“‹ ë©”ë‰´ë¥¼ ì„ íƒí•˜ì„¸ìš”:")
            print("1. ë‹¬ëŸ¬(USD) â†’ ì›í™”(KRW) ë³€í™˜")
            print("2. ì—”í™”(JPY) â†’ ì›í™”(KRW) ë³€í™˜")
            print("3. í˜„ì¬ í™˜ìœ¨ ì •ë³´ ë³´ê¸°")
            print("4. í™˜ìœ¨ ì •ë³´ ì—…ë°ì´íŠ¸")
            print("5. ì¢…ë£Œ")
            print("="*50)

            try:
                choice = input("ì„ íƒ (1-5): ").strip()

                if choice == '1':
                    self.handle_usd_conversion()
                elif choice == '2':
                    self.handle_jpy_conversion()
                elif choice == '3':
                    self.display_current_rates()
                elif choice == '4':
                    print("ğŸ”„ í™˜ìœ¨ ì •ë³´ë¥¼ ì—…ë°ì´íŠ¸í•˜ëŠ” ì¤‘...")
                    self.get_exchange_rates()
                elif choice == '5':
                    print("ğŸ‘‹ í™˜ìœ¨ ê³„ì‚°ê¸°ë¥¼ ì¢…ë£Œí•©ë‹ˆë‹¤. ê°ì‚¬í•©ë‹ˆë‹¤!")
                    break
                else:
                    print("âŒ ì˜ëª»ëœ ì„ íƒì…ë‹ˆë‹¤. 1-5 ì¤‘ì—ì„œ ì„ íƒí•´ì£¼ì„¸ìš”.")

            except KeyboardInterrupt:
                print("\nğŸ‘‹ í”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
                break
            except Exception as e:
                print(f"âŒ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")

    def handle_usd_conversion(self):
        """
        ë‹¬ëŸ¬ ë³€í™˜ ì²˜ë¦¬
        """
        try:
            print("\nğŸ’µ ë‹¬ëŸ¬(USD) â†’ ì›í™”(KRW) ë³€í™˜")
            usd_input = input("ë³€í™˜í•  ë‹¬ëŸ¬ ê¸ˆì•¡ì„ ì…ë ¥í•˜ì„¸ìš” (ì†Œìˆ˜ì  5ìë¦¬ê¹Œì§€ ì§€ì›): ")
            usd_amount = float(usd_input)

            if usd_amount < 0:
                print("âŒ ìŒìˆ˜ëŠ” ì…ë ¥í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
                return

            krw_result = self.usd_to_krw(usd_amount)
            if krw_result is not None:
                print(f"\nâœ… ë³€í™˜ ê²°ê³¼:")
                print(f"ğŸ’µ {usd_amount:,.5f} USD = ğŸ’° {krw_result:,.2f} KRW")

        except ValueError:
            print("âŒ ì˜¬ë°”ë¥¸ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        except Exception as e:
            print(f"âŒ ë³€í™˜ ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")

    def handle_jpy_conversion(self):
        """
        ì—”í™” ë³€í™˜ ì²˜ë¦¬
        """
        try:
            print("\nğŸ’´ ì—”í™”(JPY) â†’ ì›í™”(KRW) ë³€í™˜")
            jpy_input = input("ë³€í™˜í•  ì—”í™” ê¸ˆì•¡ì„ ì…ë ¥í•˜ì„¸ìš” (ì†Œìˆ˜ì  5ìë¦¬ê¹Œì§€ ì§€ì›): ")
            jpy_amount = float(jpy_input)

            if jpy_amount < 0:
                print("âŒ ìŒìˆ˜ëŠ” ì…ë ¥í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
                return

            krw_result = self.jpy_to_krw(jpy_amount)
            if krw_result is not None:
                print(f"\nâœ… ë³€í™˜ ê²°ê³¼:")
                print(f"ğŸ’´ {jpy_amount:,.5f} JPY = ğŸ’° {krw_result:,.2f} KRW")

        except ValueError:
            print("âŒ ì˜¬ë°”ë¥¸ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        except Exception as e:
            print(f"âŒ ë³€í™˜ ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {e}")

# ê°„ë‹¨ ì‚¬ìš© ì˜ˆì œ
def simple_converter():
    """
    ê°„ë‹¨í•œ ì›ë¼ì´ë„ˆ ë³€í™˜ í•¨ìˆ˜ë“¤
    """
    print("\n" + "="*50)
    print("ğŸš€ ê°„ë‹¨ ë³€í™˜ í•¨ìˆ˜ ì˜ˆì œ")
    print("="*50)

    def quick_usd_to_krw(usd_amount):
        """USDë¥¼ KRWë¡œ ë¹ ë¥´ê²Œ ë³€í™˜"""
        try:
            response = requests.get("https://api.exchangerate-api.com/v4/latest/USD")
            if response.status_code == 200:
                rate = response.json()['rates']['KRW']
                return usd_amount * rate
            return None
        except:
            return None

    def quick_jpy_to_krw(jpy_amount):
        """JPYë¥¼ KRWë¡œ ë¹ ë¥´ê²Œ ë³€í™˜"""
        try:
            response = requests.get("https://api.exchangerate-api.com/v4/latest/USD")
            if response.status_code == 200:
                rates = response.json()['rates']
                usd_amount = jpy_amount / rates['JPY']
                return usd_amount * rates['KRW']
            return None
        except:
            return None

    # ì˜ˆì œ ì‹¤í–‰
    print("ì˜ˆì œ 1: 100 USDë¥¼ KRWë¡œ ë³€í™˜")
    result = quick_usd_to_krw(100)
    if result:
        print(f"100 USD = {result:,.2f} KRW")

    print("\nì˜ˆì œ 2: 10000 JPYë¥¼ KRWë¡œ ë³€í™˜")
    result = quick_jpy_to_krw(10000)
    if result:
        print(f"10000 JPY = {result:,.2f} KRW")

# ë©”ì¸ ì‹¤í–‰ë¶€
if __name__ == "__main__":
    calculator = CurrencyCalculator()
    calculator.run_calculator()

    # ê°„ë‹¨ ë³€í™˜ í•¨ìˆ˜ ì˜ˆì œë„ ë³´ì—¬ì£¼ê¸°
    # simple_converter()
```

```python
# ExchangeRate-API ì‚¬ìš© ë²„ì „ (ë¬´ë£Œ, API í‚¤ ë¶ˆí•„ìš”)
import requests

def get_exchange_rates():
    """
    ExchangeRate-APIì—ì„œ ì‹¤ì‹œê°„ í™˜ìœ¨ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    """
    try:
        url = "https://api.exchangerate-api.com/v4/latest/USD"
        response = requests.get(url)

        if response.status_code == 200:
            return response.json()['rates']
        return None
    except Exception as e:
        print(f"API í˜¸ì¶œ ì˜¤ë¥˜: {e}")
        return None

def currency_converter():
    """
    API ê¸°ë°˜ í™˜ìœ¨ ê³„ì‚°ê¸°
    """
    print("ğŸŒ ì‹¤ì‹œê°„ í™˜ìœ¨ ê³„ì‚°ê¸° (API ê¸°ë°˜)")
    print("="*45)

    # í™˜ìœ¨ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    rates = get_exchange_rates()
    if not rates:
        print("í™˜ìœ¨ ì •ë³´ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return

    usd_krw = rates['KRW']
    jpy_usd = rates['JPY']
    jpy_krw = usd_krw / jpy_usd  # 1 JPY = ? KRW

    print(f"\nğŸ’± í˜„ì¬ í™˜ìœ¨:")
    print(f"ğŸ‡ºğŸ‡¸ 1 USD = {usd_krw:,.2f} KRW")
    print(f"ğŸ‡¯ğŸ‡µ 1 JPY = {jpy_krw:.4f} KRW")
    print(f"ğŸ‡¯ğŸ‡µ 100 JPY = {jpy_krw*100:.2f} KRW")

    while True:
        print("\n" + "="*45)
        print("1ï¸âƒ£ ë‹¬ëŸ¬ â†’ ì›í™”")
        print("2ï¸âƒ£ ì—”í™” â†’ ì›í™”")
        print("3ï¸âƒ£ í™˜ìœ¨ ì •ë³´ ìƒˆë¡œê³ ì¹¨")
        print("4ï¸âƒ£ ì¢…ë£Œ")

        choice = input("\në©”ë‰´ ì„ íƒ: ")

        if choice == '1':
            try:
                amount = float(input("ë‹¬ëŸ¬ ê¸ˆì•¡ ì…ë ¥ (ì†Œìˆ˜ì  5ìë¦¬ê¹Œì§€): "))
                result = amount * usd_krw
                print(f"\nğŸ’µ ${amount:,.5f} = ğŸ’° {result:,.2f}ì›")
            except ValueError:
                print("âŒ ìœ íš¨í•œ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”.")

        elif choice == '2':
            try:
                amount = float(input("ì—”í™” ê¸ˆì•¡ ì…ë ¥ (ì†Œìˆ˜ì  5ìë¦¬ê¹Œì§€): "))
                result = amount * jpy_krw
                print(f"\nğŸ’´ Â¥{amount:,.5f} = ğŸ’° {result:,.2f}ì›")
            except ValueError:
                print("âŒ ìœ íš¨í•œ ìˆ«ìë¥¼ ì…ë ¥í•˜ì„¸ìš”.")

        elif choice == '3':
            print("ğŸ”„ í™˜ìœ¨ ì •ë³´ ì—…ë°ì´íŠ¸ ì¤‘...")
            rates = get_exchange_rates()
            if rates:
                usd_krw = rates['KRW']
                jpy_usd = rates['JPY']
                jpy_krw = usd_krw / jpy_usd
                print("âœ… í™˜ìœ¨ ì •ë³´ê°€ ì—…ë°ì´íŠ¸ë˜ì—ˆìŠµë‹ˆë‹¤!")
                print(f"ğŸ‡ºğŸ‡¸ 1 USD = {usd_krw:,.2f} KRW")
                print(f"ğŸ‡¯ğŸ‡µ 1 JPY = {jpy_krw:.4f} KRW")
            else:
                print("âŒ í™˜ìœ¨ ì •ë³´ ì—…ë°ì´íŠ¸ ì‹¤íŒ¨")

        elif choice == '4':
            print("ğŸ‘‹ ê³„ì‚°ê¸°ë¥¼ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            break
        else:
            print("âŒ 1-4 ì¤‘ì—ì„œ ì„ íƒí•˜ì„¸ìš”.")

if __name__ == "__main__":
    currency_converter()
```
